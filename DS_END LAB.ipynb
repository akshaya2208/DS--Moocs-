```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({
    'Promo_Budget':[1000,2000,1500,3000,1200],
    'Online_Ads':[500,800,600,1200,400],
    'Footfall':[150,280,210,420,180]
})

plt.scatter(df['Promo_Budget'], df['Footfall'])
plt.xlabel("Promo Budget")
plt.ylabel("Footfall")
plt.show()

X = df[['Promo_Budget','Online_Ads']]
y = df['Footfall']
model = LinearRegression().fit(X, y)

print("Predicted Footfall:", model.predict([[2500,900]])[0])

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

pd.plotting.scatter_matrix(df)
plt.show()

plt.bar(['Promo Budget','Online Ads'], model.coef_)
plt.title("Model Coefficients")
plt.show()
```
